```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
texts = """
    «Аэрофлот» возобновил регулярные полеты в Гавану (Куба). Рейсы будут осуществляться на новейших комфортабельных самолетах Airbus A350 дважды в неделю — по вторникам и пятницам. Об этом «Ленте.ру» сообщили в пресс-службе перевозчика, где подчеркнули, что многочасовой перелет через океан будет максимально приятным и комфортным.

    Конфигурация салонов самолета Airbus A350 представлена тремя классами обслуживания для 316 пассажиров: 28 мест класса «Бизнес», 24 места класса «Комфорт», 264 места класса «Эконом». К услугам пассажиров всех классов — мультимедийная система развлечений, landscape-камеры для наблюдения за полетом и доступ в Интернет по сети Wi-Fi.

    В классе «Эконом» шаг между креслами увеличен до 78,7 сантиметра. В классе «Комфорт» пассажиры по достоинству оценят специальное меню, кресла с увеличенным углом отклонения спинки, мониторы диагональю 13 дюймов (33 сантиметра), индивидуальным освещением и розеткой для зарядки гаджетов.

    Путешественникам класса «Бизнес» предлагаются уникальные просторные кабинки SUITE с собственным мини-баром, создающие особое приватное пространство. Каждое место оборудовано монитором 21 дюйм (53 сантиметра), а кресла раскладываются в полноценную двухметровую кровать.

    Гавана стала вторым пунктом на Кубе, куда «Аэрофлот» выполняет регулярные рейсы. С июня 2021 года авиакомпания трижды в неделю летает на курорт Варадеро.
"""
```

```python
wordcloud = WordCloud().generate(text=texts)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
